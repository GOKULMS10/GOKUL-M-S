# 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

```python
#scatterplot
#Comparison of Petal length and Petal Width of each flower
sns.stripplot(x='PW',y='PL',hue='Classification',data=df)
sns.set(rc= {"figure.figsize": (10,6)})
plt.title('Comparison of Sepal length and Petal length of each flower',fontsize=21)
plt.xlabel('PETAL WIDTH')
plt.ylabel('PETAL LENGTH')
plt.show()
```

```python
#histogram of sepal length,sepal width,petal length,petal width
df.hist(color='black')
plt.show()
```

```python
#boxplot
#comparison of sepal length and Sepal width of each flower
sns.boxplot(x='SW',y='SL',data=df,hue='Classification')
plt.xlabel('SEPAL WIDTH')
plt.ylabel('SEPAL LENGTH')
plt.show()
```
