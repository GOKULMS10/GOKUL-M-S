# 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

```python
#stripplotplot
#Comparison of Petal length and Petal Width of each flower
sns.stripplot(x='PW',y='PL',hue='Classification',data=df)
sns.set(rc= {"figure.figsize": (10,6)})
plt.title('Comparison of Petal length and Petal width of each flower',fontsize=21)
plt.xlabel('PETAL WIDTH')
plt.ylabel('PETAL LENGTH')
plt.show()
```

```python
#histogram of sepal length,sepal width,petal length,petal width separately
df.hist(color='black')
plt.show()
```

```python
#boxplot
#comparison of sepal length and Sepal width of each flower
sns.boxplot(x='SW',y='SL',data=df,hue='Classification')
plt.title('Comparison of Sepal length and Sepal width of each flower',fontsize=21)
plt.xlabel('SEPAL WIDTH')
plt.ylabel('SEPAL LENGTH')
plt.show()
```

```python
# histogram of sepal length,sepal width,petal length,petal width together
df[['SL','SW','PL','PW']].plot.hist()
plt.show()
```

```python
# pairplot of all data
sns.pairplot(df)
plt.show()
```

```python
# scatterplot for comparison of sepal length and petal length
plt.figure(figsize=(10,6))
plt.scatter(df['SL'],df['PL'],s=10,c='red')
plt.title('COMPARISON OF SEPAL LENGTH AND PETAL LENGTH',fontsize=10)
plt.xlabel('SEPAL LENGTH')
plt.ylabel('PETAL LENGTH')
plt.show()
```
